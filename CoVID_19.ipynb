## COVID 19 Data- from WHO

```python
import numpy as np
import pandas as pd
import os
import urllib
%matplotlib inline
```

```python
# file path is a variable path for folder in python standard library
url = "https://covid19.who.int/WHO-COVID-19-global-table-data.csv"
file_path = os.path.join("data", "covid")
```

```python
# make directories in os filepath
os.makedirs(file_path, exist_ok=True)
csv_path = os.path.join(file_path,"WHO-COVID-19-gloval-data.csv")
urllib.request.urlretrieve(url,csv_path)
```

```python
df= pd.read_csv(csv_path)
```
